```python
import random
import time

def computer_guesses():
    """
    A game where the computer tries to guess the user's secret number.
    The user provides feedback (too high, too low, or correct) after each guess.
    """
    print("Welcome to the Reverse Guess the Number Game!")
    print("Think of a number between 1 and 100, and I'll try to guess it!")

    # Get the user's secret number
    while True:
        try:
            secret_number = int(input("Enter your secret number (1-100): "))
            if 1 <= secret_number <= 100:
                break
            else:
                print("Please enter a number between 1 and 100!")
        except ValueError:
            print("Please enter a valid number!")

    print("\nGreat! Now I'll try to guess your number...")
    time.sleep(1)  # Pause for dramatic effect

    # Initialize game variables
    low = 1
    high = 100
    attempts = 0
    max_attempts = 10  # Computer gets up to 10 guesses

    while attempts < max_attempts:
        attempts += 1

        # Computer makes a guess (using binary search for efficiency)
        guess = (low + high) // 2
        print(f"\nAttempt {attempts}: I guess {guess}!")

        # Get user feedback
        while True:
            feedback = input("Is this (C)orrect, (H)igh, or (L)ow? ").lower()
            if feedback in ['c', 'h', 'l']:
                break
            print("Please enter 'C', 'H', or 'L'!")

        # Process feedback
        if feedback == 'c':
            print(f"\nðŸŽ‰ I guessed your number {secret_number} in {attempts} attempts!")
            return
        elif feedback == 'h':
            print("Okay, I'll guess lower next time.")
            high = guess - 1
        else:
            print("Okay, I'll guess higher next time.")
            low = guess + 1

        # Check if the range is invalid (user lied)
        if low > high:
            print("\nWait a minute... that's impossible! You must have given me wrong feedback!")
            return

    print(f"\nðŸ¤– I couldn't guess your number in {max_attempts} attempts!")
    print("Either you picked a tricky number or I need better algorithms!")

# Start the game
computer_guesses()

# Option to play again
while True:
    play_again = input("\nWould you like to play again? (yes/no): ").lower()
    if play_again == 'yes':
        computer_guesses()
    elif play_again == 'no':
        print("Thanks for playing! Goodbye!")
        break
    else:
        print("Please enter 'yes' or 'no'.")
```
